```python
### Author: Md Shadman Sakib
### Equidistant Grid wind file


### things to improve:

import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
from netCDF4 import Dataset

#-------------------------------INPUT START---------------------------------------------------------------#
#start_date_and_end_date; set it according to the simulation time period required.
input_start_date = datetime(2012,1,1,0,0,0)    #yy,mm,dd,hh,min,ss
input_end_date   = datetime (2012,12,31,23,0,0)   #yy,mm,dd,hh,min,ss
f_name="Sandyera5__CPB_2012_Full"
ds=Dataset(r"G:\My Drive\Deflt3D FM Codes - Vtech\DataDownloadAPIs\ERA5\download_uvp_sandy_2012_full.nc")
factor=1.0

#-------------------------------INPUT END-----------------------------------------------------------------#
#---------------------------------------------------------------------------------------------------------#
#-------------------------------DATA PROCESSING START-----------------------------------------------------#

xll  =ds.variables['longitude'][:][0]
dx =ds.variables['longitude'][:][1]-ds.variables['longitude'][:][0]
xcol =len(ds.variables['longitude'][:])

yll  =ds.variables['latitude'][:][-1]
dy =ds.variables['latitude'][:][0]-ds.variables['latitude'][:][1]
ycol =len(ds.variables['latitude'][:])

#Write text for x direction wind
fu=open("UWIND_"+f_name+".amu","w")
fu.write("### START OF HEADER\n")
fu.write("### This file is created by Deltares\n")
fu.write("### Additional comments\n")
fu.write("FileVersion      = 1.03\n")
fu.write("filetype         =  meteo_on_equidistant_grid\n")
fu.write("NODATA_value     = -32767\n")
fu.write("n_cols           = "+str(xcol)+"\n")
fu.write("n_rows           = "+str(ycol)+"\n")
fu.write("grid_unit        = degree\n")
fu.write("x_llcenter       = "+str(xll)+"\n")
fu.write("y_llcenter       = "+str(yll)+"\n")
fu.write("dx               = "+str(dx)+"\n")
fu.write("dy               = "+str(dy)+"\n")
fu.write("n_quantity       = 1\n")
fu.write("quantity1        = x_wind\n")
fu.write("unit1            = m s-1\n")
fu.write("### END OF HEADER\n")

#Write text for v direction wind
fv=open("VWIND_"+f_name+".amv","w")
fv.write("### START OF HEADER\n")
fv.write("### This file is created by Deltares\n")
fv.write("### Additional comments\n")
fv.write("FileVersion      = 1.03\n")
fv.write("filetype         =  meteo_on_equidistant_grid\n")
fv.write("NODATA_value     = -32767\n")
fv.write("n_cols           = "+str(xcol)+"\n")
fv.write("n_rows           = "+str(ycol)+"\n")
fv.write("grid_unit        = degree\n")
fv.write("x_llcenter       = "+str(xll)+"\n")
fv.write("y_llcenter       = "+str(yll)+"\n")
fv.write("dx               = "+str(dx)+"\n")
fv.write("dy               = "+str(dy)+"\n")
fv.write("n_quantity       = 1\n")
fv.write("quantity1        = y_wind\n")
fv.write("unit1            = m s-1\n")
fv.write("### END OF HEADER\n")

#Write text for pressure
fp=open("PRESSURE_msl"+f_name+".amp","w")
fp.write("### START OF HEADER\n")
fp.write("### This file is created by Deltares\n")
fp.write("### Additional comments\n")
fp.write("FileVersion      = 1.03\n")
fp.write("filetype         =  meteo_on_equidistant_grid\n")
fp.write("NODATA_value     = -32767\n")
fp.write("n_cols           = "+str(xcol)+"\n")
fp.write("n_rows           = "+str(ycol)+"\n")
fp.write("grid_unit        = degree\n")
fp.write("x_llcenter       = "+str(xll)+"\n")
fp.write("y_llcenter       = "+str(yll)+"\n")
fp.write("dx               = "+str(dx)+"\n")
fp.write("dy               = "+str(dy)+"\n")
fp.write("n_quantity       = 1\n")
fp.write("quantity1        = air_pressure\n")
fp.write("unit1            = Pa\n")
fp.write("### END OF HEADER\n")

#Write text for precipitation
fr=open("Precipitation"+f_name+".ampr","w")
fr.write("### START OF HEADER\n")
fr.write("### This file is created by Deltares\n")
fr.write("### Additional comments\n")
fr.write("FileVersion      = 1.03\n")
fr.write("filetype         = meteo_on_equidistant_grid\n")
fr.write("NODATA_value     = -32767\n")
fr.write("n_cols           = "+str(xcol)+"\n")
fr.write("n_rows           = "+str(ycol)+"\n")
fr.write("grid_unit        = degree\n")
fr.write("x_llcenter       = "+str(xll)+"\n")
fr.write("y_llcenter       = "+str(yll)+"\n")
fr.write("dx               = "+str(dx)+"\n")
fr.write("dy               = "+str(dy)+"\n")
fr.write("n_quantity       = 1\n")
fr.write("quantity1        = precipitation\n")
fr.write("unit1            = mm/h\n")
fr.write("### END OF HEADER\n")

#time and index
base_date  = datetime.strptime(ds.variables['time'].units[12:31], '%Y-%m-%d %H:%M:%S')
data_first_date = base_date + timedelta(hours = int(ds.variables['time'][:][0]))
data_last_date  = base_date + timedelta(hours = int(ds.variables['time'][:][-1]))


if (input_start_date >= data_first_date) & (input_end_date <= data_last_date):
    
    try:
        
        #For 3 Hour interval data
        #first_time_index = int((input_start_date-data_first_date).days)*8+int(round(((input_start_date-data_first_date).seconds/3600)/3,0))
        #last_time_index  = 1+ first_time_index + int((input_end_date-input_start_date).days)*8+int(round(((input_end_date-input_start_date).seconds/3600)/3,0))
        
        #For 1 Hour interval data
        first_time_index = int((input_start_date-data_first_date).days)*24+int(round(((input_start_date-data_first_date).seconds/3600),0))
        last_time_index  = 1+ first_time_index + int((input_end_date-input_start_date).days)*24+int(round(((input_end_date-input_start_date).seconds/3600),0))
        
        
        index_dif        = last_time_index - first_time_index
        
        
        for time, time_index in zip(ds.variables['time'][first_time_index:last_time_index],np.arange(first_time_index,last_time_index,1)):
            
            time_dif = time-ds.variables['time'][:][first_time_index]
            
            time_hh     = "TIME = "+str(time_dif)+" hours since "+str(input_start_date)+" +00:00\n"
    
            fu.write(time_hh)
            fv.write(time_hh)
            fp.write(time_hh)
            fr.write(time_hh)

            
            #for lat in np.arange(len(ds.variables['latitude'][:])-1,-1,-1):
            for lat in np.arange(0,len(ds.variables['latitude'][:])):
        
                for lon in np.arange(0, len(ds.variables['longitude'][:])):
            
                    u10     = ds.variables["u10"][time_index,lat,lon]*factor
                    fu.write("\t"+str("%.3f" % u10))
            
                    v10     = ds.variables["v10"][time_index,lat,lon]*factor
                    fv.write("\t"+str("%.3f" % v10))
            
                    pre_msl = ds.variables["msl"][time_index,lat,lon]
                    fp.write("\t"+str("%.2f" % pre_msl))
                    
                    #preci = ds.variables["tp"][time_index,lat,lon]     # activate this part of precipation is needed
                    #fr.write("\t"+str("%.2f" % preci))                 # activate this part of precipation is needed
            
                fu.write("\n")
                fv.write("\n")
                fp.write("\n")
                fr.write("\n")
    
            print(time_hh)
                 
        fu.close()
        fv.close()
        fp.close()
        fr.close()
        
        
    except:
        
        print("Data is not accessed by this script") #This line is not appropriate now
        
else:
    
    print("ERROR: Given date interval is not matched with the data, please check the date and time")
    
```
