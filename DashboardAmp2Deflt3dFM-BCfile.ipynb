```python
"""
Important: How to run and use the code.
The code doesnot write the astronomical data for the first and last point in the system.
Always start the start_boundary= ##### ALWAYS AFTER THE FIRST POINT         
end_boundary=###### AT LAST POINT OF THE STATION.

It doesn't matter if you change the directory. For some reason it alwasy saves the .bc file as 
The file_name is the name of the file it is saved in the code directory.
"""



from IPython.display import HTML, display

def extract_data(input_file, output_file, start_marker, end_marker):
    # Reading the content of the astronomial constituient file
    with open(input_file, 'r') as file:
        content = file.read()
    # Extract the data between the specified markers, set according to the names of the boundary in Deflt3D Dashboard
    start_index = content.find(start_marker) + len(start_marker)
    end_index = content.find(end_marker)
    extracted_data = content[start_index:end_index].strip()
    
    
    if start_index < end_index:
        print('Extraction correct:start_index < end_index')
    else:
        #print('ERROR!!!')
        display(HTML('<b style="color:red;">ERROR!!!</b>'))

    return extracted_data

input_file_path =r"E:\VirginiaTech-Research\ChesapeakeOceanModel\Data\BnD\Files\17.5km_toolbox.bca" ###################### change her ########################
output_file_path = r"E:\VirginiaTech-Research\ChesapeakeOceanModel\Data\BnD\all astro info\Boundary_05.bc"     ###################### change her ########################

base_name = "Boundary05_"    ###################### change her ########################
boundaryside='South'     ###################### change her ########################
start_boundary=2             ###################### change her ######################## !!!!!!!!!!!!!! ALWAYS START AFTER THE FIRST POINT ########
end_boundary=16              ###################### change her ########################
file_name='Boundary_05.bc'


for bnd_code in range(start_boundary,end_boundary+1):
    start_marker = boundaryside+f"{bnd_code:02}"+'A\n'
    end_marker = boundaryside+f"{bnd_code:02}"+'B\n'
    print(base_name+f"{bnd_code:04}")
    print('Start_Marker = ', start_marker.strip())
    print('End_Marker = ', end_marker.strip())
    # extracting the data to be written in the file

    amps=extract_data(input_file_path, output_file_path, start_marker, end_marker)
    print("#####################################################")

    with open(file_name, "a") as file:
            content = f"""
[forcing]
Name                            = {base_name}{bnd_code:04}
Function                        = astronomic
Quantity                        = astronomic component
Unit                            = -
Quantity                        = waterlevelbnd amplitude
Unit                            = m
Quantity                        = waterlevelbnd phase
Unit                            = deg
    """
            
            file.write(content.strip() + "\n")
            file.write(amps + "\n\n")

print(f"Content written to {file_name}")




```
