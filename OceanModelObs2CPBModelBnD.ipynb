```python
### It will take in the pli file and make a list of the Boundary points.
### This boundary point name will be used to track down the observatio points from the his file.
### This will then be printed.
### reference time will be given as input.
### The timestep will be calculated w.r.t this reference point.
### they will be plotted aganist each other. This might not be required if we are extracting the full time scale from the ocean model and the timesapn matches
### the intended timestep we are looking to run the main model.
import pandas as pd
import numpy as np
import netCDF4 as nc



#-------------------------------FUNCTION START-------------------------------------------------------------#



#-------------------------------FUNCTION END--------------------------------------------------------------#
#---------------------------------------------------------------------------------------------------------#
#-------------------------------INPUT START---------------------------------------------------------------#
reference_time = '2001-01-01 00:00:00'

bnd_pli=pd.read_csv(r"E:\VirginiaTech-Research\ChesapeakeModel\Data\BnD\pli\Boundary03.pli", sep='\t', skiprows=2, header=None)
bnd_id=bnd_pli.iloc[:,2]
ocean_path=r"E:\VirginiaTech-Research\ChesapeakeOceanModel\SandyOceanRuns\SandyCFSv2_0.012\fm\output\SandyCFSv2_0.012_0000_his.nc"
file_name=r"E:\VirginiaTech-Research\ChesapeakeOceanModel\Data\BnD\BnDfromOceanModels\cfsv2_0.012\Boundary03.bc"


#-------------------------------INPUT END-----------------------------------------------------------------#
#---------------------------------------------------------------------------------------------------------#
#-------------------------------DATA PROCESSING START-----------------------------------------------------#

ocean_data=nc.Dataset(ocean_path)
variable = ocean_data.variables
station_array=variable['station_name']
wl=variable['waterlevel']
time=variable['time']






### iterating through all the boundary_ids
for bnd in bnd_id:
    ### iterating through all the station to find a match.
    for s in range(station_array.shape[0]):
        write_wltime_df=pd.DataFrame()
        chars=station_array[:][s].compressed()
        station=''.join([char.decode('utf-8') for char in chars]).strip()
        #print(station)
        if station==bnd:
            # extracting the boundary condition as datafarme.
            print('Found BnD Data:', bnd)
            print('------------------------------------------------')
            write_wltime_df['Seconds w.r.t Ref Time']=time[:]
            write_wltime_df['Water Level (m)']=wl[:,s]





            ### writing the dataframe into a file with extension .bc 

            with open(file_name, "a") as file:
                content = f"""
[forcing]
Name                            = {bnd}
Function                        = timeseries
Time-interpolation              = linear
Quantity                        = time
Unit                            = seconds since 2001-01-01 00:00:00
Quantity                        = waterlevelbnd
Unit                            = m
    """
    
                file.write(content.strip() + "\n")
                
                # writing out the dataframe as string 
                write_df = write_wltime_df.to_string(header=False, index=False)
                file.write(write_df + "\n")
                

            
            
```
